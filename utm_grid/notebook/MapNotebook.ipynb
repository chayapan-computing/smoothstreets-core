```python
import pygeohash as pgh 
import pandas as pd
```

```python
df = pd.read_csv('city_latlng.csv')
df
```

```python
for r in df.iterrows():
    print(r)
```

```python
n = next(df.iterrows())
n
```

```python
df.columns = ['name', 'lat', 'lng']
df
```

```python
cities = []

for k,v in df.iterrows():
    # pgh.encode()
    # print(k, h)
    lat, lng=v[1], v[2]
    h6 = pgh.encode(lat, lng, precision=6)
    h8 = pgh.encode(lat, lng, precision=8)
    print(v[0], lat, lng, h6)
    city = {'name': v[0], 'lat':lat, 'lng': lng, 'geohash_6': h6, 'geohash_8': h8}
    cities.append(city)
```

```python
df2 = pd.DataFrame(cities)
df2.to_csv('city_geohash6.csv')
```

```python
from utm import latlon_to_zone_number, from_latlon
for v in cities:
    try:
        lat, lng = v['lat'], v['lng']
        u = latlon_to_zone_number(lat, lng)
        x,y,n,z = from_latlon(lat, lng)
        print(u,lat,lng, v['name'], int(x),int(y),n,z)
    except Exception as e:
        print("error for %s %s" % (e,v))
```
